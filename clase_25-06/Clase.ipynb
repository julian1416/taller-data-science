```python
#valores faltantes de num of doors
df.replace({'num-of-doors': np.nan}, {'num-of-doors': "four"}, inplace=True)
df['num-of-doors'].head()
```

```python
# 18  bore               201 non-null    float64
# 19  stroke             201 non-null    float64
# 21  horsepower         203 non-null    float64
# 22  peak-rpm           203 non-null    float64
# 25  price              201 non-null    float64

# bore

promedio_bore = df['bore'].astype(float).mean()
print('El promedio de la columna bore: ', promedio_bore)
df.replace({'bore': np.nan}, {'bore': promedio_bore}, inplace=True)

# stroke

promedio_stroke = df['stroke'].astype(float).mean()
print('El promedio de la columna stroke: ', promedio_stroke)
df.replace({'stroke': np.nan}, {'stroke': promedio_stroke}, inplace=True)

# horsepower

promedio_horsepower = df['horsepower'].astype(float).mean()
print('El promedio de la columna horsepower: ', promedio_horsepower)
df.replace({'horsepower': np.nan}, {'horsepower': promedio_horsepower}, inplace=True)

# peak-rpm

promedio_peak_rpm = df['peak-rpm'].astype(float).mean()
print('El promedio de la columna peak-rpm: ', promedio_peak_rpm)
df.replace({'peak-rpm': np.nan}, {'peak-rpm': promedio_peak_rpm}, inplace=True)

# price

promedio_price = df['price'].astype(float).mean()
print('El promedio de la columna price: ', promedio_price)
df.replace({'price': np.nan}, {'price': promedio_price}, inplace=True)
```

```python
df_sub_set = df[["horsepower","peak-rpm", "city-mpg", "price" ]]
df_sub_set.head()
```

```python
#
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype(float)
df[['normalized-losses']] = df[['normalized-losses']].astype(int)
df[['price']] = df[['price']].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float)
```

```python
maximo = df['height'].max()
#diferenciacion
df["height"] = df["height"] / maximo
print(maximo)
df["height"]
```

```python
maximo = df['price'].max()
#diferenciacion
df["price"] = df["price"] / maximo
print(maximo)
df["price"]
```

```python
# x = (X -media )/(std)

df["peak-rpm"].head()

# Z score -3 a 3 

media = df["peak-rpm"].mean()
std_ = df["peak-rpm"].std()
df["peak-rpm"] = (df["peak-rpm"] - media)/std_
df["peak-rpm"].head()

```

```python
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
df["horsepower-binned"].value_counts()

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")

```
