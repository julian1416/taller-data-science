Consideremos la variable g y encontremos el primer índice de la subcadena snow:

```python
g = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb \
Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go"

indice_snow = g.find("snow")
print("El primer índice de 'snow' es:", indice_snow)
print("El tipo de dato es:", type(indice_snow).__name__)
```

En la variable g, reemplace la subcadena Mary por Bob:

```python
g = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb \
Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go"

g_modificada = g.replace("Mary", "Bob")

print("Reemplazando la subcadena:", g_modificada)
print("El tipo de dato:", type(g_modificada).__name__)
```

Considere la siguiente tupla:

```python
# sample tuple

genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco")
genres_tuple
```

Acceda al elemento, con respecto al índice 3:

```python
elemento_indice_3 = genres_tuple[3]

print("El elemento en el índice 3 es:", elemento_indice_3)
print("El tipo de dato es:", type(elemento_indice_3).__name__)
```

Utilice el segmentado para obtener los índices 3, 4 y 5:

```python
segmento_tupla = genres_tuple[3:6]

print("El segmento obtenido es:", segmento_tupla)
print("El tipo de dato es:", type(segmento_tupla).__name__)
```

Generar una lista ordenada a partir de la tupla C_tuple=(-5, 1, -3):

```python
C_tuple = (-5, 1, -3)

lista_ordenada = sorted(C_tuple)

print("La lista ordenada es:", lista_ordenada)
print("El tipo de dato es:", type(lista_ordenada).__name__)
```
