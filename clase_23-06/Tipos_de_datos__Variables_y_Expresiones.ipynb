¿Cuál es el tipo de dato del resultado de: 6 / 2?

```python
resultado = 6 / 2
print("El resultado es:", resultado)
print("El tipo de dato es:", type(resultado))
```

¿Qué tipo de resultado tiene la operación: 6 // 2?

```python
resultado_entero = 6 // 2
print("El resultado es:", resultado_entero)
print("El tipo de dato es:", type(resultado_entero))
```

Escribe el código para convertir el siguiente número que representa el ID de empleado "1001" a un número entero.

```python
id_texto = "1001"
resultado = int(id_texto)
print("El resultado es:", resultado)
print("El tipo de dato es:", type(resultado).__name__)
```

Escribe el código para convertir este número que representa el valor financiero "1234.56" a un número de punto flotante.

```python
valor_texto = "1234.56"
resultado_float = float(valor_texto)
print("El resultado es:", resultado_float)
print("El tipo de dato es:", type(resultado_float).__name__)
```
