## realizar una funcion que se le ingrese los coeficientes de una funcion cuadratica y retorne sus raices

```python
# ax2 + bx + c = 0
# -b +- raiz(b**2 -4 *a*c) / 2a
## x1, x2

def raiz_cero (a,b,c):
    """
    arg:
        a es numero entero
        b es numero entero
        c numero
    ouput:
        x1, x2 numeros que representan la solucion de la ecuacion
    """
    raiz = (b**2 - 4 *a*c)
    x1 = (-b+ raiz) / (2*a)
    x2 = (-b- raiz) / (2*a)
    return x1,x2
```

```python
# x2 + 2x + 1
a = 1
b = 2
c = 1

x1, x2 = raiz_cero (a,b,c)
print(f"la solucion de la ecuacion es: ")
print(f"x1 {x1}")
print(f"x2 {x2}")
```

```python
def raiz_cero (a=1,b=2,c=1):
    """
    arg:
        a es numero entero
        b es numero entero
        c numero
    ouput:
        x1, x2 numeros que representan la solucion de la ecuacion
    """
    raiz = (b**2 - 4 *a*c)
    x1 = (-b+ raiz) / (2*a)
    x2 = (-b- raiz) / (2*a)
    return x1,x2
```

```python
# x2 + 5x +6 (x+3) (x+2)
x1, x2 = raiz_cero (a = 1, b=5, c=6)

print(f"la solucion de la ecuacion es: ")
print(f"x1 {x1}")
print(f"x2 {x2}")
```
